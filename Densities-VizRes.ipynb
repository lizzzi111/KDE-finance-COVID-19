```python
import pandas as pd
import numpy as np
import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
res_pre_COVID  = pd.read_csv('data/densities_residuals_pre_covid.csv')
res_post_COVID = pd.read_csv('data/densities_residuals_post_covid.csv')
```

```python
res_pre_COVID.index = res_pre_COVID.x
res_post_COVID.index = res_post_COVID.x
```

```python
plt.figure()
ax = res_pre_COVID['dax_TVKDE'].plot(figsize=(8, 6), color='#1f77b4')
res_post_COVID['dax_TVKDE'].plot(legend=False, figsize=(8, 6), color='#42f5e3')
ax.set_xlabel('X', fontsize=20)
ax.set_ylabel('Estimated Density Function', fontsize=20)
plt.savefig('Density_DAX_res.png')
```
